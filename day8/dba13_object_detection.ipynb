```python
video_path = './data/mombi_traffic.mp4'

cap = cv2.VideoCapture(video_path)
cap.set(3,640)
cap.set(4,480)

while True:
    _, img = cap.read() # 실시간으로 화면을 캡처 ret(결과정보객체) 보통 사용하지 않아서 _로 변경, img(실시간이미지)
    # cv2.imshow('youtube mpeg', img) ## 내부적으로 PyQt로 생성되는 GUI창
    height, width = img.shape[0], img.shape[1]
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) ## 컬러 -> 흑백으로
    half = cv2.resize(gray,(int(width/2, height/2))) ## 사이즈를 반으로 축소
    cv2.imshow('youtube gray', half)

    if cv2.waitKey(1) == ord('q'): # 키보드를 q를 클릭하면
        break

cap.release() # 자원 해제
cv2.destroyAllWindows()
```

- 포토삽등의 이미지, 프리미어등의 동영상 처리하는 프로그램에서 사용하는 거의 대부분의 기능이 OpenCV에 포함되어 있음

```python
video_path = './data/news.mp4'


faceCascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(video_path) # 0~숫자는 카메라번호


while (cap.isOpened()):
    ret, img = cap.read() # 실시간으로 화면을 캡처 ret(결과정보객체) 보통 사용하지 않아서 _로 변경, img(실시간이미지)
    if ret == True:
        height, width = img.shape[0], img.shape[1]

        half = cv2.resize(img, (int(width/2), int(height/2)))

        faces = faceCascade.detectMultiScale(
            half,
            scaleFactor=1.2,
            minNeighbors=5,
            minSize=(10,10)
    )
    ## 찾은 얼굴 위치 표시
        for(x,y,w,h) in faces:
            cv2.rectangle(half(x,y),(x+w,y+h),(255,0,0),2)
            roi_color = half[y:y+h, x:x+w]

        cv2.imshow('youtube mpeg', half)

        if cv2.waitKey(1) == ord('q'): # 키보드를 q를 클릭하면
            break

cap.release() # 자원 해제
cv2.destroyAllWindows()
```

```python
video_path = './data/mombi_traffic.mp4'

cap = cv2.VideoCapture(video_path)

while (cap.isOpened()):
    ret, img = cap.read()
    if ret == True:
        height, width = img.shape[0], img.shape[1]
        half = cv2.resize(img, (int(width/2),int(height/2)))

        results = model(half, stream=True)

        for result in results:
            last = result.plot()

        cv2.imshow('YOLOv8', last)

        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

cap.release() # 자원 해제
cv2.destroyAllWindows()
```

```python
video_path = './data/mombi_traffic.mp4'

cap = cv2.VideoCapture(video_path)
out = cv2.VideoWriter('./data/mombi_traffic.mp4', fourcc=1446269005,fps=30, frameSize=(640,360))

while (cap.isOpened()):
    ret, img = cap.read()
    if ret == True:
        height, width = img.shape[0], img.shape[1]
        half = cv2.resize(img, (int(width/2),int(height/2)))
        results = model(half, stream=True)

        for result in results:
            last = result.plot()

        cv2.imshow('YOLOv8', last)
        out.write(last)

        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

cap.release() # 자원 해제
out.release()
cv2.destroyAllWindows()
```
