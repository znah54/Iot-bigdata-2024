```python
video_path = './data/news.mp4'


faceCascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(video_path) # 0~Ïà´ÏûêÎäî Ïπ¥Î©îÎùºÎ≤àÌò∏


while (cap.isOpened()):
    ret, img = cap.read() # Ïã§ÏãúÍ∞ÑÏúºÎ°ú ÌôîÎ©¥ÏùÑ Ï∫°Ï≤ò ret(Í≤∞Í≥ºÏ†ïÎ≥¥Í∞ùÏ≤¥) Î≥¥ÌÜµ ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÏïÑÏÑú _Î°ú Î≥ÄÍ≤Ω, img(Ïã§ÏãúÍ∞ÑÏù¥ÎØ∏ÏßÄ)
    if ret == True:
        height, width = img.shape[0], img.shape[1]

        half = cv2.resize(img, (int(width/2), int(height/2)))

        faces = faceCascade.detectMultiScale(
            half,
            scaleFactor=1.2,
            minNeighbors=5,
            minSize=(10,10)
    )
    ## Ï∞æÏùÄ ÏñºÍµ¥ ÏúÑÏπò ÌëúÏãú
        for(x,y,w,h) in faces:
            cv2.rectangle(half(x,y),(x+w,y+h),(255,0,0),2)
            roi_color = half[y:y+h, x:x+w]

        cv2.imshow('youtube mpeg', half)

        if cv2.waitKey(1) == ord('q'): # ÌÇ§Î≥¥ÎìúÎ•º qÎ•º ÌÅ¥Î¶≠ÌïòÎ©¥
            break

cap.release() # ÏûêÏõê Ìï¥Ï†ú
cv2.destroyAllWindows()
```

```python
video_path = './data/mombi_traffic.mp4'

cap = cv2.VideoCapture(video_path)
out = cv2.VideoWriter('./data/mombi_traffic.mp4', fourcc=1446269005,fps=30, frameSize=(640,360))

while (cap.isOpened()):
    ret, img = cap.read()
    if ret == True:
        height, width = img.shape[0], img.shape[1]
        half = cv2.resize(img, (int(width/2),int(height/2)))
        results = model(half, stream=True)

        for result in results:
            last = result.plot()

        cv2.imshow('YOLOv8', last)
        out.write(last)

        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

cap.release() # ÏûêÏõê Ìï¥Ï†ú
out.release()
cv2.destroyAllWindows()
```
