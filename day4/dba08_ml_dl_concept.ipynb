### 머신러닝, 딥러닝 기초

- 인공지능(Artificial Intelligence) - 기계가 사람의 지능을 모방하는 기술

### 머신러닝, 딥러닝 라이브러리

#### Pandas
- 데이터 과학분야 사용되는 오픈소스, 라이브러리 데이터 조작, 정리 등의 핵심도구

#### Numpy
- 과학연산에 초점을 맞춘 오픈소스 라이브러리, 행렬 계산에 특화, 난수 생성등

#### Scipy
- Numpy 확장 라이브러리, 과학컴퓨팅, 데이터처리, 고성능 컴퓨팅 오픈소스 라이브러리

#### Scikit Learn
- 파이썬 대표적인 머신러닝 라이브러리(딥러닝은 없음)
- 지도 및 비지도 학습 알고리즘 개발, 적용
- Numpy, SciPy, Matplotlib 패키지와 통합되어 머신러닝 프로젝트에 필수적으로 사용
- 회귀, 분류, 차원축소, 클러스터링(군집) 등 기능 포함 가장 최초에 머신러닝을 배울 때 많이 사용
- 쉽게 빼울 수 있음

#### Theano
- 수학계산, 식정의 최적화 라이브러리 중 하나, 대규모 다차원 배열 처리
- NumPy와 통합, GPU 사용가능, 미분이 쉽고, 속도와 안정성 최적화, C코드 생성
- 딥러닝 연구 및 개발에 업계 표준으로 사용된다?

#### Tensorflow
- 구글에서 만든 고성능 수치 계산 라이브러리 머신러닝, 딥러닝 알고리즘 포함
- 훈련, 개발하는 데 널리 사용되는 오픈소스 라이브러리
- 2015년 발표, 파이썬, C++, CUDA로 개발
- 딥러닝 분야를 거의 장악했었음
- 구글 코랩에서 바로 사용
- 2.0으로 버전업 되면서 Keras를 포함한 1.x와 2.x는 상이함


#### PyTorch
- 메타 AI(페이스북)에서 만든 머신러닝, 딥러닝 라이브러리
- 2016년 10월 발표, 파이썬, C++, CUDA로 개발
- 텐서플로우 비직관적인 구조와 난이도 때문에 사용자가 늘고 있는 추세
- 연구개발에는 파이토치가 더 많이 사용되고 있음
- 2022년 기준으로 텐서플로 사용률 넘어섬
- 영상처리 분야 Yolo 라이브러리가 기반임
