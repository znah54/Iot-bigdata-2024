- 웹크롤링 가능여부 판단
    - 웹사이트 열고, 페이지 소스 보기를 했을때 HTML소스가 일반적으로 표시되어야 함
    - 네이버 등 대형포털은 크롤링 방지가 되어 있음 

- 뷰티플수프 사용
- requests 인터넷 모듈도 같이 사용

````python
from bs4 import BeautifulSoup  # 줄임말 보통안씀
import requests
````

````python
## 크롤링 사이트 멜론 : https://www.melon.com/chart/index.htm
## 406 에러 : 클라이언트 오류 응답 코드
## 크롤링 방지 사이트 : 다나와, 구글, 네이버, 
url = 'https://www.bluer.co.kr/search?query=%EC%86%8C%EB%A3%A1%ED%8F%AC&isSearchName=false&location=' # 여기를 크로링 할 것

response = requests.get(url)

if response.status_code == 200: # 200 OK, 404 Not Found, 500 Internal Error
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup)
    # soup.select()
else:
    print(response.status_code)
````

- 단순 웹크롤링은 현재 대부분 사이트가 막고 있음
